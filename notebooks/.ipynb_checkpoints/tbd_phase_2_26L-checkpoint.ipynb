## Generator template

The helper below creates a common base event table. You should extend it for your variant.

Do not spend most of the assignment writing a perfect data generator. The generator only needs to create data that is large enough and structurally interesting enough for your benchmark questions.

```python
def skewed_ids(rng, n, max_id, hot_fraction=0.02, hot_probability=0.50):
    hot_count = max(1, int(max_id * hot_fraction))
    ids = rng.integers(hot_count + 1, max_id + 1, size=n)
    hot_mask = rng.random(n) < hot_probability
    ids[hot_mask] = rng.integers(1, hot_count + 1, size=hot_mask.sum())
    return ids


def random_tag_lists(rng, n, vocabulary=None, min_tags=1, max_tags=3):
    vocabulary = np.array(vocabulary or ["ai", "cloud", "spark", "polars", "duckdb", "sql", "etl", "security", "mlops"])
    counts = rng.integers(min_tags, max_tags + 1, size=n)
    tag_ids = rng.integers(0, len(vocabulary), size=(n, max_tags))
    return [[str(vocabulary[tag_ids[i, j]]) for j in range(counts[i])] for i in range(n)]


def generate_base_events(n, rng):
    start = np.datetime64("2026-01-01T00:00:00", "s")
    end = np.datetime64("2026-04-01T00:00:00", "s")
    seconds = int((end - start) / np.timedelta64(1, "s"))
    event_ts = (start + rng.integers(0, seconds, size=n).astype("timedelta64[s]")).astype("datetime64[us]")

    df = pl.DataFrame(
        {
            "event_id": np.arange(1, n + 1),
            "entity_id": skewed_ids(rng, n, max_id=200_000),
            "event_ts": event_ts,
            "category": rng.choice(["A", "B", "C", "D", "E", "F"], size=n),
            "country": rng.choice(["PL", "DE", "FR", "UK", "US", "IN", "BR"], size=n),
            "device": rng.choice(["mobile", "desktop", "tablet"], size=n, p=[0.65, 0.25, 0.10]),
            "metric_1": rng.lognormal(mean=4.0, sigma=1.0, size=n).round(3),
            "metric_2": rng.integers(0, 10_000, size=n),
            "tags": random_tag_lists(rng, n),
        }
    )
    return df.with_columns(pl.col("event_ts").dt.date().alias("event_date"))


def customize_for_variant(df, card, rng):
    n = df.height
    
    # Generate weather metrics
    temperatures = rng.normal(loc=15.0, scale=10.0, size=n).round(1)
    humidities = rng.integers(20, 101, size=n)
    
    # Create masks for missing values (15% temp, 20% humidity)
    temp_null_mask = rng.random(n) < 0.15
    hum_null_mask = rng.random(n) < 0.20
    
    # Adapt columns to weather domain
    df = df.rename({
        "entity_id": "station_id",
        "device": "sensor_type",
        "category": "station_type"
    }).with_columns([
        pl.Series("temperature", temperatures),
        pl.Series("humidity", humidities),
        pl.Series("temp_null_mask", temp_null_mask),
        pl.Series("hum_null_mask", hum_null_mask)
    ])
    
    # Apply nulls based on masks
    df = df.with_columns([
        pl.when(pl.col("temp_null_mask")).then(None).otherwise(pl.col("temperature")).alias("temperature"),
        pl.when(pl.col("hum_null_mask")).then(None).otherwise(pl.col("humidity")).alias("humidity")
    ]).drop(["metric_1", "metric_2", "temp_null_mask", "hum_null_mask"])
    
    return df


def generate_dimension_table(card, rng):
    n_stations = 200_000
    
    return pl.DataFrame({
        "station_id": np.arange(1, n_stations + 1),
        "climate_zone": rng.choice(["Tropical", "Dry", "Temperate", "Continental", "Polar"], size=n_stations, p=[0.05, 0.15, 0.50, 0.25, 0.05]),
        "elevation_meters": rng.exponential(scale=300, size=n_stations).round(0),
        "maintenance_priority": rng.choice(["High", "Medium", "Low"], size=n_stations, p=[0.1, 0.3, 0.6])
    })
```

```python
# Generate and save the dataset
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

base_events = generate_base_events(N_ROWS, rng)
events = customize_for_variant(base_events, CARD, rng)
dimension = generate_dimension_table(CARD, rng)

events.write_parquet(EVENTS_PATH, compression="zstd")
dimension.write_parquet(DIMENSION_PATH, compression="zstd")

# Optional partitioned layout for experiments with predicate pushdown and file layout.
events.write_parquet(PARTITIONED_EVENTS_DIR, partition_by="event_date", compression="zstd")

events.sort(["event_date", "country"]).write_parquet(
    OPTIMIZED_EVENTS_PATH,
    compression="zstd",
    row_group_size=100_000,
)
events.drop("tags").write_csv(CSV_EVENTS_PATH)

manifest = {
    "created_at_utc": datetime.now(timezone.utc).isoformat(),
    "group_id": GROUP_ID,
    "variant": CARD,
    "scale": SCALE,
    "rows": int(events.height),
    "run_seed": RUN_SEED,
    "paths": {
        "events": str(EVENTS_PATH),
        "events_partitioned": str(PARTITIONED_EVENTS_DIR),
        "events_optimized": str(OPTIMIZED_EVENTS_PATH),
        "dimension": str(DIMENSION_PATH),
    },
    "environment": {
        "python": platform.python_version(),
        "polars": pl.__version__,
        "pandas": pd.__version__,
        "duckdb": duckdb.__version__,
        "cpu_logical_cores": psutil.cpu_count(logical=True),
        "ram_gib": round(psutil.virtual_memory().total / 2**30, 2),
    },
}
MANIFEST_PATH.write_text(json.dumps(manifest, indent=2), encoding="utf-8")

print(json.dumps(manifest, indent=2))
```

## Part 2: Measuring performance

You must use one consistent benchmark protocol for all libraries/engines.

Minimum requirements:

1. Run every benchmark at least three times. Five repetitions are recommended.
2. Run `gc.collect()` before each measured repetition to reduce noise from previous Python allocations.
3. Report median runtime, not only one measurement.
4. Record peak memory where possible.
5. Check that results are logically equivalent across libraries/engines.
6. Store your results in a table.
7. Describe any library/engine-specific settings, such as Pandas dtype backend, thread count, Spark local mode, or DuckDB threads.

**Important for memory benchmarks**: notebook kernels keep allocations and library state between cells. Peak-RSS comparisons are often misleading when all variants run in the same process. For Task 3.1 and any memory-sensitive comparison, prefer running each variant in a fresh process or a small standalone script. If you cannot do that, clearly state this limitation.

You may use the helper shape below, but you need to implement the actual benchmark functions.


```python
BENCHMARK_COLUMNS = [
    "library_engine",
    "mode",
    "query_name",
    "data_format",
    "layout",
    "rows",
    "median_time_s",
    "peak_memory_mb",
    "input_size_mb",
    "result_check",
    "notes",
]

# Suggested output: one dictionary matching BENCHMARK_COLUMNS per library/engine/query/mode.

benchmark_results = []

def get_path_size_mb(path):
    """Calculates the size of a file or directory (e.g., partitioned Parquet) in MB."""
    path_obj = Path(path)
    if not path_obj.exists():
        return 0.0
    if path_obj.is_file():
        return path_obj.stat().st_size / (1024 * 1024)
    return sum(f.stat().st_size for f in path_obj.glob('**/*') if f.is_file()) / (1024 * 1024)

def run_benchmark(query_name, library_engine, mode, query_func, data_path, layout="default", data_format="parquet", repetitions=5, notes=""):
    """
    Helper function to run benchmarks.
    Measures execution time (median of `repetitions` runs) and peak memory usage.
    """
    times = []
    
    # 1. Time measurements (with gc.collect() before each iteration)
    for _ in range(repetitions):
        gc.collect()
        start = time.perf_counter()
        _ = query_func()
        times.append(time.perf_counter() - start)
        
    median_time_s = np.median(times)
    
    # 2. Memory measurement (run the function once more with the profiler)
    gc.collect()
    # Get baseline memory usage of the process before running the function
    baseline_mem = memory_usage(-1, interval=0.1, timeout=0.1)[0]
    
    # Run the function and measure peak memory usage (also returns the function's result)
    mem_usage, result = memory_usage((query_func,), max_usage=True, retval=True)
    
    # Peak usage is the difference between max and baseline (safeguard against negative values)
    peak_memory_mb = max(0.0, mem_usage - baseline_mem)
    
    # 3. Input data size
    input_size_mb = get_path_size_mb(data_path) if data_path else 0.0
    
    # 4. Attempt to extract the number of rows from the result (depends on the library)
    try:
        if isinstance(result, list): # PySpark (collect)
            rows = len(result)
        elif hasattr(result, "shape"): # Pandas, DuckDB (df)
            rows = result.shape[0]
        elif hasattr(result, "height"): # Polars
            rows = result.height
        elif hasattr(result, "count"): # PySpark (if DataFrame)
            rows = result.count()
        else:
            rows = len(result)
    except Exception:
        rows = None

    # 5. Save results
    record = {
        "library_engine": library_engine,
        "mode": mode,
        "query_name": query_name,
        "data_format": data_format,
        "layout": layout,
        "rows": rows,
        "median_time_s": round(median_time_s, 4),
        "peak_memory_mb": round(peak_memory_mb, 2),
        "input_size_mb": round(input_size_mb, 2),
        "result_check": "Pending", # For manual verification later
        "notes": notes
    }
    
    benchmark_results.append(record)
    print(f"[{library_engine} - {mode}] {query_name}: Time = {median_time_s:.4f}s, Peak Mem = {peak_memory_mb:.2f} MB")
    
    return result
```

**Memory Measurement Limitation Note:**
We measure peak memory using `memory_profiler` by tracking the OS-level RSS of the Jupyter process and subtracting the baseline memory before execution. Because all engines run in the same notebook kernel, cached allocations (e.g., memory pools in Polars or DuckDB) might not be fully released to the OS between cells. We mitigate this by calling `gc.collect()`, but the peak memory values should be treated as estimates of *additional* memory required by the query, rather than absolute standalone process limits.

### Query 1: Selective Filter & Aggregation (Pruning Sensitive)
**Description:** Calculate the average temperature and humidity for a specific country and a specific sensor type, ignoring null values.
**Classes covered:** Selective filter plus aggregation, sensitive to column/row-group pruning.
**Hypothesis:**
- *What does it test?* Predicate pushdown (filtering by country/sensor) and column pruning (only reading 4 columns). It also tests basic null handling in aggregations.
- *Best engine?* DuckDB or Polars (lazy) because they will push the filters down to the Parquet reader and skip irrelevant row groups.
- *Most memory?* Pandas (default backend), as it will likely load the entire dataset into memory before applying the filter.
- *Which layout helps?* The optimized Parquet layout (sorted by country and date) will significantly speed this up via row-group pruning.

### Query 2: Dimension Join & High-Cardinality Group-By
**Description:** Join the `events` table with the `dimension` table on `station_id`. Group by `climate_zone` and `station_type`, and calculate the maximum temperature and the count of missing (null) humidity readings.
**Classes covered:** Join with a dimension table, high-cardinality group-by.
**Hypothesis:**
- *What does it test?* Hash join performance between a large fact table and a dimension table, followed by a multi-column aggregation.
- *Best engine?* DuckDB, as it has a highly optimized vectorized execution engine for joins and aggregations.
- *Most memory?* Pandas, because merging large DataFrames materializes massive intermediate objects in memory.
- *Which layout helps?* Physical layout of the fact table won't help much here since we need to scan almost all rows to perform the join.

### Query 3: Window Computation & Top-K (Null Handling Stress)
**Description:** For each `station_id`, find the top 3 highest temperature readings. This requires filtering out null temperatures, partitioning by `station_id`, ordering by `temperature` descending, and taking the top 3.
**Classes covered:** Window computation, top-k, high-cardinality group-by.
**Hypothesis:**
- *What does it test?* Window functions (or optimized top-k algorithms) over a very high number of groups (200,000 stations), while correctly ignoring nulls.
- *Best engine?* Polars or DuckDB. Polars has highly optimized `.top_k_by()` and window expressions.
- *Most memory?* PySpark (local) or Pandas. Window functions in Spark can cause memory pressure if partitions are skewed or too large for local memory.
- *Which layout helps?* Sorting the dataset by `station_id` and `temperature` beforehand would make this query trivial, but on the default layout, it requires heavy shuffling/sorting.

```python
def run_pandas_benchmarks(backend_name, use_pyarrow=False):
    kwargs = {"engine": "pyarrow", "dtype_backend": "pyarrow"} if use_pyarrow else {}
    
    def q1():
        df = pd.read_parquet(EVENTS_PATH, **kwargs)
        filtered = df[(df['country'] == 'PL') & (df['sensor_type'] == 'mobile')]
        return filtered[['temperature', 'humidity']].mean()

    def q2():
        df = pd.read_parquet(EVENTS_PATH, **kwargs)
        dim = pd.read_parquet(DIMENSION_PATH, **kwargs)
        merged = df.merge(dim, on='station_id')
        merged['hum_is_null'] = merged['humidity'].isna()
        return merged.groupby(['climate_zone', 'station_type']).agg(
            max_temp=('temperature', 'max'),
            null_hum_count=('hum_is_null', 'sum')
        )

    def q3():
        df = pd.read_parquet(EVENTS_PATH, **kwargs)
        return df.dropna(subset=['temperature']) \
                 .sort_values('temperature', ascending=False) \
                 .groupby('station_id').head(3)

    print(f"\n--- Running Pandas ({backend_name}) ---")
    run_benchmark("Q1: Filter & Agg", "Pandas", backend_name, q1, EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "Pandas", backend_name, q2, EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "Pandas", backend_name, q3, EVENTS_PATH)

run_pandas_benchmarks("default", use_pyarrow=False)
run_pandas_benchmarks("pyarrow", use_pyarrow=True)
```

```python
def run_polars_benchmarks():
    # --- EAGER ---
    def q1_eager():
        df = pl.read_parquet(EVENTS_PATH)
        return df.filter((pl.col('country') == 'PL') & (pl.col('sensor_type') == 'mobile')) \
                 .select([pl.col('temperature').mean(), pl.col('humidity').mean()])

    def q2_eager():
        df = pl.read_parquet(EVENTS_PATH)
        dim = pl.read_parquet(DIMENSION_PATH)
        return df.join(dim, on='station_id') \
                 .group_by(['climate_zone', 'station_type']) \
                 .agg([pl.col('temperature').max().alias('max_temp'), 
                       pl.col('humidity').is_null().sum().alias('null_hum_count')])

    def q3_eager():
        df = pl.read_parquet(EVENTS_PATH)
        return df.drop_nulls('temperature') \
                 .sort(['station_id', 'temperature'], descending=[False, True]) \
                 .group_by('station_id', maintain_order=True).head(3)

    # --- LAZY (Default & Streaming) ---
    def get_q1_lazy(streaming=False):
        def _q1():
            q = pl.scan_parquet(EVENTS_PATH) \
                  .filter((pl.col('country') == 'PL') & (pl.col('sensor_type') == 'mobile')) \
                  .select([pl.col('temperature').mean(), pl.col('humidity').mean()])
            return q.collect(engine="streaming") if streaming else q.collect()
        return _q1

    def get_q2_lazy(streaming=False):
        def _q2():
            q = pl.scan_parquet(EVENTS_PATH).join(pl.scan_parquet(DIMENSION_PATH), on='station_id') \
                  .group_by(['climate_zone', 'station_type']) \
                  .agg([pl.col('temperature').max().alias('max_temp'), 
                        pl.col('humidity').is_null().sum().alias('null_hum_count')])
            return q.collect(engine="streaming") if streaming else q.collect()
        return _q2

    def get_q3_lazy(streaming=False):
        def _q3():
            q = pl.scan_parquet(EVENTS_PATH).drop_nulls('temperature') \
                  .sort(['station_id', 'temperature'], descending=[False, True]) \
                  .group_by('station_id').head(3)
            return q.collect(engine="streaming") if streaming else q.collect()
        return _q3

    print("\n--- Running Polars ---")
    # Eager
    run_benchmark("Q1: Filter & Agg", "Polars", "eager", q1_eager, EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "Polars", "eager", q2_eager, EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "Polars", "eager", q3_eager, EVENTS_PATH)
    
    # Lazy Default
    run_benchmark("Q1: Filter & Agg", "Polars", "lazy", get_q1_lazy(False), EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "Polars", "lazy", get_q2_lazy(False), EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "Polars", "lazy", get_q3_lazy(False), EVENTS_PATH)
    
    # Lazy Streaming
    run_benchmark("Q1: Filter & Agg", "Polars", "streaming", get_q1_lazy(True), EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "Polars", "streaming", get_q2_lazy(True), EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "Polars", "streaming", get_q3_lazy(True), EVENTS_PATH)

run_polars_benchmarks()
```

```python
def run_duckdb_benchmarks():
    events_p = f"'{EVENTS_PATH}'"
    dim_p = f"'{DIMENSION_PATH}'"
    
    def q1():
        query = f"""
            SELECT AVG(temperature), AVG(humidity) 
            FROM read_parquet({events_p}) 
            WHERE country = 'PL' AND sensor_type = 'mobile'
        """
        return duckdb.sql(query).df()

    def q2():
        query = f"""
            SELECT d.climate_zone, e.station_type, 
                   MAX(e.temperature) as max_temp, 
                   SUM(CASE WHEN e.humidity IS NULL THEN 1 ELSE 0 END) as null_hum_count 
            FROM read_parquet({events_p}) e 
            JOIN read_parquet({dim_p}) d ON e.station_id = d.station_id 
            GROUP BY d.climate_zone, e.station_type
        """
        return duckdb.sql(query).df()

    def q3():
        query = f"""
            WITH ranked AS (
                SELECT station_id, temperature, 
                       ROW_NUMBER() OVER(PARTITION BY station_id ORDER BY temperature DESC) as rn 
                FROM read_parquet({events_p}) 
                WHERE temperature IS NOT NULL
            ) 
            SELECT station_id, temperature 
            FROM ranked 
            WHERE rn <= 3
        """
        return duckdb.sql(query).df()

    print("\n--- Running DuckDB ---")
    run_benchmark("Q1: Filter & Agg", "DuckDB", "sql", q1, EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "DuckDB", "sql", q2, EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "DuckDB", "sql", q3, EVENTS_PATH)

run_duckdb_benchmarks()
```

```python
def run_pyspark_benchmarks():
    def q1():
        df = spark.read.parquet(str(EVENTS_PATH))
        return df.filter("country = 'PL' AND sensor_type = 'mobile'") \
                 .agg(F.avg("temperature"), F.avg("humidity")) \
                 .collect()

    def q2():
        events = spark.read.parquet(str(EVENTS_PATH))
        dim = spark.read.parquet(str(DIMENSION_PATH))
        return events.join(dim, "station_id") \
                     .groupBy("climate_zone", "station_type") \
                     .agg(F.max("temperature").alias("max_temp"), 
                          F.sum(F.when(F.col("humidity").isNull(), 1).otherwise(0)).alias("null_hum_count")) \
                     .collect()

    def q3():
        events = spark.read.parquet(str(EVENTS_PATH))
        window = Window.partitionBy("station_id").orderBy(F.col("temperature").desc())
        return events.filter(F.col("temperature").isNotNull()) \
                     .withColumn("rn", F.row_number().over(window)) \
                     .filter(F.col("rn") <= 3) \
                     .drop("rn") \
                     .collect()

    print("\n--- Running PySpark (Local) ---")
    run_benchmark("Q1: Filter & Agg", "PySpark", "local", q1, EVENTS_PATH)
    run_benchmark("Q2: Join & GroupBy", "PySpark", "local", q2, EVENTS_PATH)
    run_benchmark("Q3: Window Top-K", "PySpark", "local", q3, EVENTS_PATH)

run_pyspark_benchmarks()
```

**Selected Query:** Query 1 (Filter by `country` and `sensor_type`, aggregate `temperature` and `humidity`).

**Layout Strategy & Hypothesis:**
1. **Default Parquet:** Data is randomly ordered. The engine can perform column pruning (reading only 4 out of all columns), but row-group pruning is ineffective because 'PL' and 'mobile' records are scattered across all row groups.
2. **Optimized Parquet:** We sort the data by `country` and `sensor_type` and reduce the `row_group_size` to 100,000. 
   * *Why it helps:* Parquet stores min/max statistics for each row group. By sorting, we cluster all 'PL' and 'mobile' records together. The engine can check the min/max stats in the Parquet footer and completely skip (prune) row groups that do not contain our target values, drastically reducing disk I/O.
3. **CSV (Negative Baseline):** A flat text file. 
   * *Why it's worse:* CSV has no metadata and no columnar structure. The engine must read the entire file from disk, parse every row, and read all columns just to evaluate the filter.
